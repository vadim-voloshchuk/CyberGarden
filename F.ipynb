```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
langs = open('langs.txt', 'r').read().split('\n')
langs = langs[26:]
langs
```

```python
import time

def get_hubs_from_lang(lang):
    name = []
    description = []
    keys = []
    last_data = []
    license = []
    language = []
    stars = []

    for page in range(25):
        time.sleep(5)
        res = requests.get(f'https://github.com/search?l={lang}&o=desc&p={page+1}&q={lang}&s=stars&type=Repositories')
        soup = BeautifulSoup(res.text, "html.parser")
        print(f"{lang} {page + 1}")
        allNReps = soup.findAll('li', class_='repo-list-item')
        
        if allNReps == []:
            break
            
        for rep in allNReps:
            lowers = rep.findAll('div', class_='mr-3')
            
            if lowers[0].text.replace('  ', '').replace('\n', '')[0].isdigit()  == False:
                  break
            
            tags = rep.findAll('a', class_='mx-0')
            if len(tags) > 0:
                keys.append(','.join([tag.text.replace('  ', '').replace('\n', '') for tag in tags]))
            else: 
                keys.append('None')
            print(lowers[0].text.replace('  ', '').replace('\n', '')[0])
                  
            name.append(rep.find('a', class_='v-align-middle').text)
            description.append(rep.find('p', class_='mb-1').text if rep.find('p', class_='mb-1') != None else "None")
            
            stars.append(float(lowers[0].text.replace('  ', '').replace('\n', '').replace('k', '000').replace('.', '')))
            language.append(lowers[1].text.replace('  ', '').replace('\n', ''))


            if len(lowers) == 3:
                license.append('None')
                last_data.append(lowers[2].text.replace('  ', '').replace('\n', ''))

            else:
                license.append(lowers[2].text.replace('  ', '').replace('\n', ''))
                last_data.append(lowers[3].text.replace('  ', '').replace('\n', ''))

    new_data = pd.DataFrame({"name": name, "description": description, "keys": keys, "last_data": last_data, "license": license, "language": language, "stars": stars})
    time.sleep(30)

    return new_data
```

```python
lang_data_full = pd.DataFrame()

for lang in langs:
    lang_data = get_hubs_from_lang(lang)
    
    lang_data_full = pd.concat([lang_data_full, lang_data])

    
    
```
